```python
import numpy as np
import cv2
import glob
import utils
import matplotlib.pyplot as plt
from lane_finder import LaneFinder
from moviepy.editor import VideoFileClip
%matplotlib inline

images = glob.glob('./camera_cal/calibration*.jpg')
mtx, dist = utils.calibrate_camera(images)

img = utils.load_image('./test_images/test5.jpg')

lane_finder = LaneFinder(mtx, dist
                        )
warped = lane_finder.create_mask(img)
plt.imshow(warped, cmap='gray')
```

```python
left_center, right_center = utils.get_line_centers(warped)
print(left_center, right_center)

lefty, leftx = utils.get_lane(warped, left_center)
righty, rightx = utils.get_lane(warped, right_center)

left_fit = utils.get_line_coef(leftx, lefty)
right_fit = utils.get_line_coef(rightx, righty)

y = np.linspace(0, warped.shape[0]-1, warped.shape[0] )
left_x = utils.fit_line(left_fit, y)
right_x = utils.fit_line(right_fit, y)

out_img = np.zeros_like(img)
out_img[lefty, leftx] = [255, 0, 0]
out_img[righty, rightx] = [0, 0, 255]

plt.imshow(out_img)
plt.plot(left_x, y, color='yellow')
plt.plot(right_x, y, color='yellow')
plt.xlim(0, 1280)
plt.ylim(720, 0)
```

```python
result = lane_finder.project_on_image(img, left_x, right_x, y)
plt.imshow(result)
```
